```python
# 導入必要的庫和模塊
import os
import leafmap
from samgeo import SamGeo, tms_to_geotiff, get_basemaps
```

```python
# 使用您提供的座標創建地圖對象
m = leafmap.Map(center=[19.0384, 72.8488], zoom=19)
m.add_basemap("SATELLITE")  # 添加衛星圖層作為底圖

# 獲取用戶感興趣的區域或設置默認範圍
if m.user_roi_bounds() is not None:
    bbox = m.user_roi_bounds()
else:
    bbox = [72.8483, 19.0383, 72.8493, 19.0393]  # 根據提供的座標設置默認邊界框

# 定義輸出圖像和分割圖像的文件名
image = "satellite.tif"
mask = "segment.tif"
vector = "segment.gpkg"
shapefile = "segment.shp"

# 從選定來源下載並轉換TMS圖磚到GeoTIFF
tms_to_geotiff(output=image, bbox=bbox, zoom=20, source="Satellite", overwrite=True)

# 初始化SamGeo模型
sam = SamGeo(model_type="vit_h", checkpoint="sam_vit_h_4b8939.pth")

# 生成分割掩碼並將其保存
sam.generate(image, mask, batch=True, foreground=True, erosion_kernel=(3, 3), mask_multiplier=255)

# 將分割掩碼轉換為GeoPackage和Shapefile格式
sam.tiff_to_gpkg(mask, vector, simplify_tolerance=None)
sam.tiff_to_vector(mask, shapefile)

# 為矢量圖層設置樣式
style = {
    "color": "#3388ff",
    "weight": 2,
    "fillColor": "#7c4185",
    "fillOpacity": 0.5,
}

# 在地圖上添加矢量圖層
m.add_vector(vector, layer_name="Vector", style=style)

# 顯示地圖
m
```
